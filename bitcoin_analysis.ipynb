# Bitcoin Price Analysis

# Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load Dataset

```python
df = pd.read_csv("BTC-USD.csv")
```

## Data Inspection

### Preview Data


```python
df.head()
```

### Dataset Information

```python
df.info()
```

### Dataset Shape


```python
print(f"Rows and Columns: {df.shape}")
```

### Dataset Columns

```python
print(df.columns)
```

## Data Cleaning

### Missing Values

```python
df.isnull().sum()
```

### Duplicates

```python
df.duplicated().sum()
```

### Convert Date

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df.info()
```

### Sort Data

```python
df = df.sort_values("Date")
```

### Visualize Bitcoin Price

```python
plt.figure(figsize=(10,5))

plt.plot(df["Date"], df["Close"])

plt.title("Bitcoin Price")
plt.xlabel("Date")
plt.ylabel("Price")

plt.show()
```

### Calculate Daily Returns

```python
df["Daily_Return"] = df["Close"].pct_change()
```

### Visualize Daily Returns

```python
plt.figure(figsize=(10,5))

plt.plot(df["Date"], df["Daily_Return"])

plt.title("Daily Returns")
plt.xlabel("Date")
plt.ylabel("Returns")

plt.show()
```

## Descriptive Statistics

### Mean

```python
close_mean = df["Close"].mean()
print(f"close mean :{close_mean}")
```

```python
plt.figure(figsize=(10,5))

plt.plot(df["Date"], df["Close"])
plt.axhline(
    y=close_mean,
    linestyle="--",
    color="red",
    label="Mean"
)

plt.title("Price Means")
plt.xlabel("Date")
plt.ylabel("Returns")

plt.show()
```

### Median

```python
close_median = df["Close"].median()

print(f"close mean :{close_median}")
```

### Mode

```python
mode_Close = df["Close"].mode()

print(f"mode_Close: {mode_Close}")
```

### Min and Max

```python
close_min = df["Close"].min()
close_max = df["Close"].max()

print(f"Minimum Close Price: {close_min}")
print(f"Maximum Close Price: {close_max}")
```

### Range

```python
close_range = close_max - close_min

print(f"Price Range: {close_range}")
```
